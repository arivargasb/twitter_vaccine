```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import json
```

```python
tweets = pd.read_csv('vaccine_us_0720to0121.csv')
tweets.head()
```

```python
# check size of the dataframe
tweets.shape
```

```python
# check for NaN values
tweets.isna().sum()
```

```python
# check for strange values
tweets['date'].unique()
```

```python
# create a copy to clean
tweets_clean = tweets.copy()

# check shape
tweets_clean.shape
```

```python
# drop rows with all NaN values
tweets_clean.dropna(how='all', inplace=True)

# confirm
tweets_clean.shape
```

```python
# drop duplicate tweets and/or retweets
tweets_clean = tweets_clean.drop_duplicates(subset='tweet')

# confirm
tweets_clean.shape
```

```python
# take another look at NaN values
tweets_clean.isna().sum()
```

```python
# drop any tweets missing a value for name (takes care of other NaN values)
tweets_clean.dropna(subset=['name'], inplace=True)
tweets_clean.shape
```

```python
# drop unwanted columns
tweets_clean.drop(columns=['time', 'timezone', 'place' , 'retweet_id',
                          'language', 'mentions', 'urls', 'photos',
                          'cashtags', 'quote_url', 'video', 'thumbnail',
                          'near', 'geo', 'source', 'user_rt_id',
                          'user_rt', 'retweet_id', 'retweet_date', 'translate',
                          'trans_src', 'trans_dest'], inplace=True)
```

```python
tweets_clean
```

```python
# reset index
tweets_clean.reset_index(drop=True, inplace=True)

# confirm
tweets_clean.head()
```

```python
# save to csv
tweets_clean.to_csv('tweets_clean.csv')
```
